```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
n_samples = 200


X1 = np.random.normal(50, 10, n_samples)
X2 = X1 + np.random.normal(0, 5, n_samples)  # Correlated with X1
X3 = X1 * 0.5 + np.random.normal(0, 3, n_samples)  # Correlated with X1
X4 = np.random.normal(60, 12, n_samples)  # Less correlated with X1
X5 = X2 + X3 * 0.3 + np.random.normal(0, 7, n_samples)  # Mixed correlation

df_corr = pd.DataFrame({
    'Feature 1': X1,
    'Feature 2': X2,
    'Feature 3': X3,
    'Feature 4': X4,
    'Feature 5': X5
})

# Calculate the correlation matrix
corr_matrix = df_corr.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#4 using x and y data generate a 3d surface plot to visualize the function
import numpy as np
import plotly.graph_objects as go

# Define the grid of x and y values
x = np.linspace(-5, 5, 50)  # 50 points between -5 and 5
y = np.linspace(-5, 5, 50)

# Create a meshgrid for x and y
x, y = np.meshgrid(x, y)

# Define the function z = f(x, y)
z = np.sin(np.sqrt(x**2 + y**2))  # Example: radial sine wave

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Customize the layout
fig.update_layout(
    title="3D Surface Plot of sin(sqrt(x^2 + y^2))",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

# Display the plot
fig.show()
```

```python
#5 create a bubble chart to represent each country population( y axis), gdp(x axis) and bubble size proportional  to population
import plotly.express as px
import pandas as pd

# Sample data: Country, Population, GDP
data = {
    'Country': ['USA', 'China', 'India', 'Germany', 'UK'],
    'Population': [331, 1439, 1380, 83, 68],  # in millions
    'GDP': [21137518, 14722731, 2875142, 3845630, 2827113]  # in billion USD
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bubble chart
fig = px.scatter(df, x='GDP', y='Population',
                 size='Population',  # Bubble size based on population
                 hover_name='Country',  # Display country name on hover
                 size_max=100,  # Max size for the bubbles
                 title='Country Population vs GDP')

# Customize axes labels
fig.update_layout(
    xaxis_title="GDP (in Billion USD)",
    yaxis_title="Population (in Millions)"
)

# Show the chart
fig.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh output in the notebook
output_notebook()

# Define x and y values
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=600, height=400)

# Add a line to the plot
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Enable Bokeh output in the notebook
output_notebook()

# Generate random data for x and y values
np.random.seed(42)
x = np.random.rand(100) * 10  # 100 random x values between 0 and 10
y = np.random.rand(100) * 10  # 100 random y values between 0 and 10

# Generate random sizes and colors for markers
sizes = np.random.randint(10, 50, 100)  # Random sizes between 10 and 50
colors = np.random.randint(50, 255, 100)  # Random values for colors between 50 and 255

# Create a DataFrame to hold the data
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a linear color mapper
color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=min(df['colors']), high=max(df['colors']))

# Create a new plot
p = figure(title="Scatter Plot with Random Data", x_axis_label='X Values', y_axis_label='Y Values', width=600, height=400)

# Add a scatter plot to the figure
p.scatter(x='x', y='y', size='sizes', color=color_mapper, source=source, fill_alpha=0.6, legend_label='Random Points')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Enable Bokeh output in the notebook
output_notebook()

# Data: Fruit names and their counts
fruits = ['Apples', 'Bananas', 'Grapes', 'Oranges', 'Peaches', 'Strawberries']
counts = [10, 25, 13, 20, 8, 15]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="", width=600, height=400)

# Add bar glyphs
p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
# Create a Bokeh histogram to visualize the distribution of the given data.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
from bokeh.models import ColumnDataSource

# Enable Bokeh output in the notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
data = np.random.randn(1000)  # 1000 random values from a standard normal distribution

# Create histogram data using NumPy's histogram function
hist, edges = np.histogram(data, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=600, height=400)

# Add quad glyphs to create the histogram
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color="blue", line_color="white", alpha=0.6)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
from bokeh.models import ColumnDataSource

# Enable Bokeh output in the notebook
output_notebook()

# Generate data
data_heatmap = np.random.randn(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data and meshgrid for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600)

# Define the color mapper
color_mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Add squares to the plot
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=color_mapper, line_color=None)

# Show the plot
show(p)
```
